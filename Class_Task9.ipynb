```python
class TourPlanner:
  def __init__(self, Company_name):
    self.Company_name = Company_name

    self.tour_package = [
        {'name': 'Gujrat', 'Price': 10000},
        {'name': 'Rajasthan', 'Price': 15000},
        {'name': 'Sikkim', 'Price': 20000}]

  def book(self, person_name, number_of_people, package_name):
    for tour in self.tour_package:
      if tour['name'] == package_name:
        total_amount = tour['Price'] * number_of_people
        return self.generate_bill(person_name, number_of_people, package_name, total_amount)

    return "sorry! Package not available."

  def generate_bill(self, person_name, number_of_people, package_name, total_amount):
    return (
        f"-------- TOUR BILL --------\n"
        f"Travel Company : {self.Company_name}\n"
        f"Customer name  : {person_name}\n"
        f"People count   : {number_of_people}\n"
        f"Tour Package   : {package_name}\n"
        f"Total amount   : ₹ {total_amount}\n"
        f"---------------------------\n"
    )


Planner = TourPlanner("Easy_Trip")
print(Planner.book("Rohit Sharma", 2, "Gujrat"))

Planner = TourPlanner("Dream_Traveller")
print(Planner.book("Virat Kohli", 5, "Maysoor"))
```
